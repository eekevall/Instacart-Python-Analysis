## 03. Data Wrangling and Consistency Checks on orders_products_customers_merged

```python
# Drop the Unnamed: 0, Unnamed: 0.1 and _merge columns which are not required for this analysis and inclusion could breach data ethics
df_analysis.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1', '_merge'], inplace = True)
df_analysis
```

### Check NaN values and change data types

```python
# Count NaN values in the dataframe
df_analysis.isnull().sum()
```

### Change data types

```python
# Change the data types of the 'user_id' to string, and 'gender', 'US_state', 'no_dependents' 'family_status' to category
df_analysis = df_analysis.astype({'user_id': str, 'gender': 'category', 'US_state': 'category', 'no_dependents': 'int16', 'family_status': 'category'})
```

```python
# Change the 'date_joined' data type  to datetime
df_analysis['date_joined'] = pd.to_datetime(df_analysis['date_joined'])
```

```python
# Change the age and income data types to lower storage capacities 
df_analysis['age'] = df_analysis['age'].astype('int16')
df_analysis['income'] = df_analysis['income'].astype('int32')
```

```python
# Change the data types of the 'order_id', 'user_id', 'product_id', 'aisle_id' and 'department_id' columns to string (these variables are qualitative nominal) and reordered to boolean (this is a yes/no variable))
df_analysis = df_analysis.astype({'order_id': str, 'product_id': str, 'aisle_id': str, 'department_id': str, 'reordered': bool})
```

```python
# Change the data types of the 'loyalty_flag', 'spending_flag', and 'frequency_flag' to cartegory (these values are ordinal categories)
df_analysis = df_analysis.astype({'price_range_loc': 'category', 'busiest_day': 'category', 'Busiest_days': 'category', 'busiest_period_of_the_day': 'category', 'loyalty_flag': 'category',  'spending_flag': 'category', 'frequency_flag': 'category'})
```

```python
# Change the data types of the 'days_since_prior_order' and 'customer_median_prior_order_days' columns to Int32
df_analysis = df_analysis.astype({'order_number': 'int16', 'orders_day_of_week': 'int16', 'order_hour_of_day': 'int16', 'add_to_cart_order': 'int16', 'max_order': 'int16'})
```

```python
# Change the data types of the 'days_since_prior_order' and 'customer_median_prior_order_days' columns to Int16
df_analysis = df_analysis.astype({'days_since_prior_order': 'Int16', 'customer_median_prior_order_days': 'Int16'})
```

```python
# Show the data types in the data frame and check the memory usage
df_analysis.info()
```

## 06. Stakeholder Questions

3. You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the â€œorder_hour_of_dayâ€ column.
In a markdown cell beneath your histogram, describe what the histogram demonstrates.

```python
# Create a histogram to show orders by hour of the day - 24 hours therefore 24 bins
hist_order_hour_of_day = sns.histplot(df_analysis, x = 'order_hour_of_day', bins = 24)
sns.despine(top = True, right = True)
plt.ticklabel_format(style = 'plain')
plt.xlabel('Hour of day', fontsize = 13, fontweight = 'semibold')
plt.ylabel('Order frequency', fontsize = 13, fontweight = 'semibold')
plt.title('Frequency of Products Ordered During the Day', fontsize = 15, fontweight = 'bold')
plt.xlim = (0, 23)
plt.xticks([0, 5, 10, 15, 20], fontsize = 11)
plt.ylim = (0, 3000000)
plt.yticks([0, 500000, 1000000, 1500000, 2000000, 2500000, 3000000], fontsize = 11)
plt.show()
```

**The Frequency of Products Ordered During the Day histogram shows that most of the orders are placed between 9 am and 4 pm (around 2.5 million orders per hour). This peak tails off around 5 pm and there are fewer orders (below 5 million orders) between 11 pm and 6 am.**

```python
# Export the Frequency of Products Ordered During the Day histogram in the form of an image file and view
hist_order_hour_of_day.figure.savefig(os.path.join(path, '04_Analysis','Visualisations', 'frq_order_hour_of_day.png'))
hist_order_hour_of_day.figure
```
