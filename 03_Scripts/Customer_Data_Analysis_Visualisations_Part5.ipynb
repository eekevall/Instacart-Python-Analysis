## 03. Data Wrangling and Consistency Checks on analysis_active_customers_profiles

```python
# Count NaN values in the dataframe
df_analysis_active_quest.isnull().sum()
```

### Change data types

```python
# Change the data types of the 'user_id' to string, and 'gender', 'US_state', 'no_dependents' 'family_status' to category
df_analysis_active_quest = df_analysis_active_quest.astype({'user_id': str, 'gender': 'category', 'US_state': 'category', 'no_dependents': 'int16', 'family_status': 'category'})
```

```python
# Change the 'date_joined' data type  to datetime
df_analysis_active_quest['date_joined'] = pd.to_datetime(df_analysis_active_quest['date_joined'])
```

```python
# Change the age and income data types to lower storage capacities 
df_analysis_active_quest['age'] = df_analysis_active_quest['age'].astype('int16')
df_analysis_active_quest['income'] = df_analysis_active_quest['income'].astype('int32')
```

```python
# Change the data types of the 'loyalty_flag', 'spending_flag', and 'frequency_flag' to cartegory (these values are ordinal categories)
ddf_analysis_active_quest = df_analysis_active_quest.astype({'price_range_loc': 'category', 'busiest_day': 'category', 'Busiest_days': 'category', 'busiest_period_of_the_day': 'category', 'loyalty_flag': 'category',  'spending_flag': 'category', 'frequency_flag': 'category'})
```

```python
# Change the data types of the 'days_since_prior_order' and 'customer_median_prior_order_days' columns to Int32
df_analysis_active_quest = df_analysis_active_quest.astype({'order_number': 'int16', 'orders_day_of_week': 'int16', 'order_hour_of_day': 'int16', 'add_to_cart_order': 'int16', 'max_order': 'int16'})
```

```python
# Change the data types of the 'days_since_prior_order' and 'customer_median_prior_order_days' columns to Int16
df_analysis_active_quest = ddf_analysis_active_quest.astype({'days_since_prior_order': 'Int16', 'customer_median_prior_order_days': 'Int16'})
```

```python
# Change the data types of the 'user_id' to string, and 'gender', 'US_state', 'no_dependents' 'family_status' to category
df_analysis_active_quest = df_analysis_active_quest.astype({'exclusion_flag': 'category', 'generation_flag': 'category', 'income_flag': 'category', 'diet_flag': 'category', 'profile': 'category'})
```

1. The sales team needs to know what the busiest days of the week and hours of the day are (i.e., the days and times with the most orders) in order to schedule ads at times when there are fewer orders.

### Busiest days of week and hours of the day

```python
# Create a bar chart frequency of orders during the week by index order with colour

# define data and colour palette
data = df_analysis_active_quest['orders_day_of_week'].value_counts().sort_index()
color = ['cadetblue', 'cadetblue', 'darkgrey', 'darkgrey', 'darkgrey', 'darkgrey', 'darkgrey']
bar_orders_day_week_kq = data.plot.bar(color = color)

# set plot layout
plt.tight_layout()
bar_orders_day_week_kq.spines['top'].set_visible(False)
bar_orders_day_week_kq.spines['right'].set_visible(False)

# set titles and axes
plt.ticklabel_format(axis="y", style='plain')
plt.xlabel('Day of the week', fontsize=13, fontweight = 'bold')
plt.ylabel('Order frequency', fontsize=13, fontweight = 'bold')
plt.xticks(rotation=0)
plt.title('Frequency of Products Ordered During the Week', fontsize=15, fontweight = 'bold')

plt.show()
```

**The bar chart shows that the weekend, Sat (day 0) and Sun (day 1), are the busiest days of the week. the least busy days are around the middle of the week Tue (day 2) and Wed (day 3).**

```python
# Export the bar chart frequency of orders during the week in the form of an image file.
bar_orders_day_week_kq.figure.savefig(os.path.join(path, '04_Analysis','Visualisations', 'bar_orders_day_week_kq.png'), bbox_inches='tight')
bar_orders_day_week_kq.figure
```

```python
# Create a histogram to show orders by hour of the day - 24 hours therefore 24 bins

# define data and colour palette
hist_order_hour_of_day_kq = sns.histplot(df_analysis_active_quest, x = 'order_hour_of_day', color = ['cadetblue'], bins = 24)

# set plot layout
plt.tight_layout()
sns.despine(top = True, right = True)

# set titles and axes
plt.ticklabel_format(style = 'plain')
plt.xlabel('Hour of day', fontsize = 13, fontweight = 'semibold')
plt.ylabel('Order frequency', fontsize = 13, fontweight = 'semibold')
plt.title('Frequency of Products Ordered During the Day', fontsize = 15, fontweight = 'bold')
plt.xlim = (0, 23)
plt.xticks([0, 5, 10, 15, 20], fontsize = 11)
plt.ylim = (0, 3000000)
plt.yticks([0, 500000, 1000000, 1500000, 2000000, 2500000, 3000000], fontsize = 11)

plt.show()
```

**The Frequency of Products Ordered During the Day histogram shows that most of the orders are placed between 9 am and 4 pm (around 2.5 million orders per hour). This peak tails off around 5 pm and there are fewer orders (below 5 million orders) between 11 pm and 6 am.**

```python
# Export the Frequency of Products Ordered During the Day histogram in the form of an image file and view
hist_order_hour_of_day_kq.figure.savefig(os.path.join(path, '04_Analysis','Visualisations', 'frq_order_hour_of_day_kq.png'), bbox_inches='tight')
hist_order_hour_of_day_kq.figure
```
