```python
# Count NaN values in the dataframe
df_analysis_active_profiles.isnull().sum()
```

### Change data types

```python
# Change the data types of the 'user_id' to string, and 'gender', 'US_state', 'no_dependents' 'family_status' to category
df_analysis_active_profiles = df_analysis_active_profiles.astype({'user_id': str, 'gender': 'category', 'US_state': 'category', 'no_dependents': 'int16', 'family_status': 'category'})
```

```python
# Change the 'date_joined' data type  to datetime
df_analysis_active_profiles['date_joined'] = pd.to_datetime(df_analysis_active_profiles['date_joined'])
```

```python
# Change the age and income data types to lower storage capacities 
df_analysis_active_profiles['age'] = df_analysis_active_profiles['age'].astype('int16')
df_analysis_active_profiles['income'] = df_analysis_active_profiles['income'].astype('int32')
```

```python
# Change the data types of the 'order_id', 'product_id', 'aisle_id' and 'department_id' columns to string (these variables are qualitative nominal) and reordered to boolean (this is a yes/no variable))
df_analysis_active_profiles = df_analysis_active_profiles.astype({'order_id': str, 'product_id': str, 'aisle_id': str, 'department_id': str})
```

```python
# Change the data types of the 'loyalty_flag', 'spending_flag', and 'frequency_flag' to cartegory (these values are ordinal categories)
df_analysis_active_profiles = df_analysis_active_profiles.astype({'price_range_loc': 'category', 'busiest_day': 'category', 'Busiest_days': 'category', 'busiest_period_of_the_day': 'category', 'loyalty_flag': 'category',  'spending_flag': 'category', 'frequency_flag': 'category'})
```

```python
# Change the data types of the 'days_since_prior_order' and 'customer_median_prior_order_days' columns to Int32
df_analysis_active_profiles = df_analysis_active_profiles.astype({'order_number': 'int16', 'orders_day_of_week': 'int16', 'order_hour_of_day': 'int16', 'add_to_cart_order': 'int16', 'max_order': 'int16'})
```

```python
# Change the data types of the 'days_since_prior_order' and 'customer_median_prior_order_days' columns to Int16
df_analysis_active_profiles = df_analysis_active_profiles.astype({'days_since_prior_order': 'Int16', 'customer_median_prior_order_days': 'Int16'})
```

```python
# Change the data types of the 'user_id' to string, and 'gender', 'US_state', 'no_dependents' 'family_status' to category
df_analysis_active_profiles = df_analysis_active_profiles.astype({'exclusion_flag': 'category', 'generation_flag': 'category', 'income_flag': 'category', 'diet_flag': 'category', 'profile': 'category'})
```
